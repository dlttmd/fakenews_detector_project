```python
import requests
import pandas as pd
```

```python
def search_news(keyword):
  url = "https://openapi.naver.com/v1/search/news.json"
  header = {
      "X-Naver-Client-Id":"발급받은 api 아이디",
      "X-Naver-Client-Secret":"발급받은 api secret"
  }
  params = {
      "query":keyword,
      "display":50
  }

  response = requests.get(url, headers = header, params = params)

  data = response.json()

  news = []
  for item in data['items']:
    title = item.get('title', '')
    description = item.get('description', '')

    news.append({'Title': title, 'Description': description})

  df = pd.DataFrame(news)

  return df
```

```python
df = search_news("손흥민")
```
